# Contexto

Dados: Série Histórica de Preços de Combustíveis e de GLP
Data de criação: 17/08/2023
Fonte: https://dados.gov.br/dados/conjuntos-dados/serie-historica-de-precos-de-combustiveis-e-de-glp

# Extract

```python
import pandas as pd

df = pd.read_csv('Preços semestrais - AUTOMOTIVOS_2023.01.csv',sep=';')
df.head()
```

# Transform

```python
df.columns = df.columns.str.replace(' ','_')
df['Valor_de_Venda'] = df['Valor_de_Venda'].str.replace(',','.')
df['Valor_de_Venda'] = df['Valor_de_Venda'].astype(float)
data_coleta_inicio = df['Data_da_Coleta'].min()
data_coleta_fim = df['Data_da_Coleta'].max()
produto_mais_vendido = df.Produto.value_counts().index[0]
qtd_mais_vendido = df.Produto.value_counts().max()
pr_medio_venda = df['Valor_de_Venda'].mean()
```

# Load

```python
print('Qual foi o período da coleta dos dados?')
print(f'>>> A coleta dos dados foi realizada de {data_coleta_inicio} até {data_coleta_fim}.')
print()
print('Qual foi o produto mais vendido?')
print(f'>>> O produto mais vendido foi {produto_mais_vendido}, com {qtd_mais_vendido} vezes.')
print()
print('Qual o preço médio de venda?')
print(f'>>> O preço médido de venda é de R$ {pr_medio_venda:,.4f}.')
```
